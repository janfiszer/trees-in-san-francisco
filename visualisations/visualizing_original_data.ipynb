## Plotting unmodified data
Visualizing provided data is very valuable for creating data processing procedures.
In this notebook we try to see whether/how these two datasets align.

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import osmnx
```

### Loading data

```python
trees_gdf = gpd.read_file("../data/SF_Urban_Tree_Canopy.geojson")
trees_gdf.head()
```

### Plotting trees

```python
trees_gdf.plot(figsize=(20,20), color="green")
plt.title("Trees in San Francisco")
```

### Loading street data

```python
osm_tags = {
    'highway': [
    "motorway",
    "trunk",
    "primary",
    "secondary",
    "living_street",
    "primary_link",
    "service",
    "unclassified",
    "motorway_link",
    "secondary_link",
    "tertiary",
    "residential",
    "trunk_link",
    "tertiary_link",
    "cycleway",
    "path",
    "track"
]
}
streets_gdf = osmnx.geometries_from_place("San Francisco", tags=osm_tags)
```

### Loading streets data

```python
streets_gdf.plot(figsize=(20,20), color="gray")
plt.title("Chosen streets in San Francisco")
```

### Plotting trees mapped onto streets

```python
# clearing all the pyplot data
plt.cla()
plt.clf()

# creating new subplot
fig, ax = plt.subplots(figsize=(20,20))
ax.set_aspect('equal')

# making trees appear on top of the streets
trees_gdf.plot(ax=ax, color='green', zorder=2)

# making streets appear under trees
streets_gdf.plot(ax=ax, color='grey', zorder=1)

plt.title("Trees mapped onto streets in San Francisco")
plt.show()
```
